```python
import seaborn as sns

import matplotlib.pyplot as plt

sns.set(style="whitegrid") # opções: darkgrid, whitegrid, dark, white, ticks

df_tips = sns.load_dataset('tips')

print(df_tips)
df_tips.drop_duplicates(keep='last', inplace=True)
```

```python
fig, ax = plt.subplots(1, 4, figsize=(10, 5))   # SUBPLOTS Definem a quantidade de graficos e altura e largura

sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[0])  # Essa função define o nome das coluna, com base nas informações disponiveis... Male, Female ) x = sexy
                                                              # Essa função distribui a média dos gastos da coluna com base nas informações automaticas da função )  y = distribuição do total dos gastos pelo eixo
                                                              # A definição da média é definida pela função, apenas por não informar nenhum parametro a função já retorna a média

sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[1], estimator=sum) # Essa função soma os valores  dos valores do eixo y

sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[2], estimator=len) # Essa função busca quantos itens tem dentro do container de 0 à 244 por categoria sexy
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

df = sns.load_dataset('tips')

plt.figure(figsize=(8, 5))
sns.barplot(x='time', y='total_bill', data=df, estimator=sum, errorbar=None, hue='time', legend=False, palette="Set2")
plt.xlabel('Período (Time)')
plt.ylabel('Total de Gastos')
plt.title('Total de Gastos por Período (Almoço ou Jantar)')
plt.show()
```

Já conseguimos verificar que o gasto no jantar é bem superior ao gasto no almoço. Mas quantos clientes almoçaram e quantas clientes jantaram? Vamos investigar o gasto médio por período.

```python
plt.figure(figsize=(8, 5))

sns.barplot(x='time', y='total_bill', data=df)#, #estimator=sum, ci=None, palette=“Set2”)

plt.xlabel('Período (Time)')

plt.ylabel('Média de Gastos')

plt.title('Média de Gastos por Período (Almoço ou Jantar)')

plt.show()
```

Entendemos que a média gasta no jantar também é superior à do almoço. Logo, esse é o principal período do restaurante em relação a faturamento.

Por fim, vamos verificar a média de gorjeta por período.

```python
# Crie um gráfico de barras com o Seaborn para mostrar a média de gorjetas por período

plt.figure(figsize=(8, 5))

sns.barplot(x='time', y='tip', data=df, hue='time', legend=False, palette="Set3")

plt.xlabel('Período (Time)')

plt.ylabel('Média da Gorjeta')

plt.title('Média da Gorjeta por Período (Almoço ou Jantar)')

plt.show()
```
