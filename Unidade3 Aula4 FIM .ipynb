```python
import seaborn as sns

import matplotlib.pyplot as plt

sns.set(style="whitegrid") # opes: darkgrid, whitegrid, dark, white, ticks

df_tips = sns.load_dataset('tips')

print(df_tips)
df_tips.drop_duplicates(keep='last', inplace=True)
```

```python
fig, ax = plt.subplots(1, 4, figsize=(10, 5))   # SUBPLOTS Definem a quantidade de graficos e altura e largura

sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[0])  # Essa funo define o nome das coluna, com base nas informaes disponiveis... Male, Female ) x = sexy
                                                              # Essa funo distribui a mdia dos gastos da coluna com base nas informaes automaticas da funo )  y = distribuio do total dos gastos pelo eixo
                                                              # A definio da mdia  definida pela funo, apenas por no informar nenhum parametro a funo j retorna a mdia

sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[1], estimator=sum) # Essa funo soma os valores  dos valores do eixo y

sns.barplot(data=df_tips, x='sex', y='total_bill', ax=ax[2], estimator=len) # Essa funo busca quantos itens tem dentro do container de 0  244 por categoria sexy
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")

df = sns.load_dataset('tips')

plt.figure(figsize=(8, 5))
sns.barplot(x='time', y='total_bill', data=df, estimator=sum, errorbar=None, hue='time', legend=False, palette="Set2")
plt.xlabel('Perodo (Time)')
plt.ylabel('Total de Gastos')
plt.title('Total de Gastos por Perodo (Almoo ou Jantar)')
plt.show()
```

J conseguimos verificar que o gasto no jantar  bem superior ao gasto no almoo. Mas quantos clientes almoaram e quantas clientes jantaram? Vamos investigar o gasto mdio por perodo.

```python
plt.figure(figsize=(8, 5))

sns.barplot(x='time', y='total_bill', data=df)#, #estimator=sum, ci=None, palette=Set2)

plt.xlabel('Perodo (Time)')

plt.ylabel('Mdia de Gastos')

plt.title('Mdia de Gastos por Perodo (Almoo ou Jantar)')

plt.show()
```

Entendemos que a mdia gasta no jantar tambm  superior  do almoo. Logo, esse  o principal perodo do restaurante em relao a faturamento.

Por fim, vamos verificar a mdia de gorjeta por perodo.

```python
# Crie um grfico de barras com o Seaborn para mostrar a mdia de gorjetas por perodo

plt.figure(figsize=(8, 5))

sns.barplot(x='time', y='tip', data=df, hue='time', legend=False, palette="Set3")

plt.xlabel('Perodo (Time)')

plt.ylabel('Mdia da Gorjeta')

plt.title('Mdia da Gorjeta por Perodo (Almoo ou Jantar)')

plt.show()
```
