✅ Iniciar um Banco de Dados - importação de biblioteca " sqlite3 "

```python
import sqlite3
# 1. Conectar ao banco de dados (ou criar um novo)
Conn = sqlite3.connect('Vandão.db')
# 2. Criar um objeto cursor
cursor = Conn.cursor()

# 3. Definir o comando SQL para criar a tabela
create_table = """CREATE TABLE IF NOT EXISTS Produtos (

    id INTEGER PRIMARY KEY,

    nome TEXT NOT NULL,

    preco REAL NOT NULL,

    estoque INTEGER

);"""
# 4. Executar o comando SQL para criar a tabela
cursor.execute(create_table)

# 5. Confirmar as alterações (commit)
Conn.commit()
# 6. Fechar a conexão com o banco de dados
Conn.close()
```

✅ Update - Atualizar um produto

```python
import sqlite3

# Conectando ao banco de dados
conn = sqlite3.connect('Vandão.db')
cursor = conn.cursor()

# Novo preço e ID do produto a ser atualizado
novo_preco = 80.99     # novo_preço - nome do comando que adicionou o produto
produto_id = 2     # Suponha que queiramos atualizar o produto com ID 1

# Comando SQL para atualizar o preço do produto
atualizar_preco = "UPDATE Produtos SET preco = ? WHERE id = ?"
# Executando o comando SQL de atualização
cursor.execute(atualizar_preco, (novo_preco, produto_id))

# Confirmando as alterações
conn.commit()
# Fechando a conexão
conn.close()
```

✅ Delete - Excluir produto da tabela

```python
import sqlite3

# Conectando ao banco de dados
conn = sqlite3.connect('Vandão.db')
cursor = conn.cursor()

# ID do produto a ser excluído
produto_id = 4  # Suponha que queiramos excluir o produto com ID 2
# Comando SQL para excluir o produto
excluir_produto = "DELETE FROM Produtos WHERE id = ?"
# Executando o comando SQL de exclusão
cursor.execute(excluir_produto, (produto_id,))

# Confirmando as alterações
conn.commit()
# Fechando a conexão
conn.close()
```

✅ APLICANDO A AULA

Suponha agora que você precise criar uma tabela de
contatos para comunicação da empresa que você
trabalha. Nessa tabela "Contatos" precisa armazenar
informações de contatos, incluindo nome, e-mail e
número de telefone.

```python
import sqlite3

# CREATE (Criação da tabela e inserção de dados de exemplo)

conn = sqlite3.connect('Vandão.db')
cursor = conn.cursor()
cursor.execute('''

    CREATE TABLE IF NOT EXISTS Contatos (

        id INTEGER PRIMARY KEY AUTOINCREMENT,

        nome TEXT,

        email TEXT,

        telefone TEXT

    )

''')

# INSERT ( Inserir os registros )
registro_clientes = [

    ('João', 'joao@email.com', '123-456-7890'),

    ('Maria', 'maria@email.com', '987-654-3210'),

    ('Carlos', 'carlos@email.com', '555-555-5555')

]

# Comando para inserir na tabela "Contatos", usar o many no final de execute para adicionar varios registros
cursor.executemany('INSERT INTO Contatos (nome, email, telefone) VALUES (?, ?, ?)', registro_clientes)
# Confirmando as alterações
conn.commit()

# Comando para fazer READ (Leitura e exibição dos registros). Obs: * asteristico mostra tudo
cursor.execute('SELECT * FROM Contatos')  # Obs: * asteristico mostra tudo que está na tabela

clientes = cursor.fetchall() # para recuperar todas as linhas do resultado de uma consulta

print("registros:")

for registro in clientes:

    print(registro)

# UPDATE (Atualização do número de telefone do contato com ID 2)
novo_telefone = '999-999-9999'
registro_id = 2
cursor.execute('UPDATE Contatos SET telefone = ? WHERE id = ?', (novo_telefone, registro_id))
conn.commit()

# DELETE (Exclusão do contato com ID 1)

registro_id_para_excluir = 1
cursor.execute('DELETE FROM Contatos WHERE id = ?', (registro_id_para_excluir,))
conn.commit()

# Fechando a conexão

conn.close()
```
