```python
import pandas as pd
import sqlite3

# Conectar ao banco de dados
conn = sqlite3.connect('dados_vendas.db')

# Ler dados da tabela vendas1 para um DataFrame
df_vendas1 = pd.read_sql_query("SELECT * FROM vendas1", conn)

# Exibir informações sobre o DataFrame
print(df_vendas1.info())

# Fechar a conexão com o banco de dados
conn.close()
vendas1=df_vendas1
print(vendas1)
```

```python
from datetime import date
from datetime import datetime as dt
data_extracao = date.today()
df_vendas1['data_extracao'] = data_extracao
df_vendas1['Responsavel'] = "Evandro Nunes"
print(df_vendas1.info())
df_vendas1.head(14)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15, 3))

data_venda = vendas1['data_venda']
valor_venda = vendas1['valor_venda']

# Criar um gráfico de barras
plt.bar(data_venda, valor_venda, color='green')
# Adicionar rótulos aos eixos
plt.xlabel('Data da Venda')
plt.ylabel('Valor da Venda R$')
# Adicionar um título ao gráfico
plt.title('Balanço')
# Mostrar o gráfico
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
plt.figure(figsize=(15, 6))
sns.barplot(x='data_venda', y='valor_venda', data=df_vendas1, estimator=sum, errorbar=None, palette="Set2")
plt.xlabel('Período (Data)')
plt.ylabel('Valor Total de Venda R$')
plt.title('Balanço das vendas (todos os periodos)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Crie um gráfico de barras com o Seaborn para mostrar a média por período
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 6))
sns.barplot(x='data_venda', y='valor_venda', data=df_vendas1, estimator='mean', errorbar=None, palette="Set3")
plt.xlabel('Data da Venda')
plt.ylabel('Média do Valor de Venda R$')
plt.title('Média do Valor de Venda por Data')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
